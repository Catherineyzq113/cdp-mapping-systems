```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

```

```python
# Sample DataFrame
data = {
    'Rent': [3000, 3500, 5000, 4500, 6000],
    'Story_Level': [5, 10, 20, 15, 25],
    'View_Quality': [6, 8, 9, 7, 10]
}

df = pd.DataFrame(data)

# 3D Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['Story_Level'], df['View_Quality'], df['Rent'], c=df['View_Quality'], cmap='viridis', s=100)

ax.set_xlabel('Story Level')
ax.set_ylabel('View Quality')
ax.set_zlabel('Rent (USD)')
ax.set_title('3D Plot of Rent, Story Level, and View Quality')

plt.show()
```

```python
import pandas as pd

# Example data
data = {
    'Rent': [3000, 3500, 5000, 4500, 6000, 7000, 8000, 7500, 8200, 9200],
    'Story_Level': [5, 10, 20, 15, 25, 30, 35, 40, 45, 50],
    'View_Quality': [6, 8, 9, 7, 10, 9, 10, 9, 10, 9]
}

df = pd.DataFrame(data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a pivot table for heatmap
pivot_table = df.pivot_table(values='Rent', index='Story_Level', columns='View_Quality', aggfunc='mean')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='.0f')
plt.title('Heatmap of Rent by Story Level and View Quality')
plt.xlabel('View Quality')
plt.ylabel('Story Level')
plt.show()
```
